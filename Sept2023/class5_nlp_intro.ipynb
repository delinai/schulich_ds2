### An Example with a movie review data set

This data set contains movie reviews and sentiment denoted at "pos" for positive review or "neg" for negative review. The idea is we can train a model to classify reviews as positive or negative, which can quickly tell us how well a movie did. 

```python
import nltk
import random
from nltk.corpus import movie_reviews
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Download the required datasets
nltk.download('movie_reviews')

# Step 1: Load Data
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]

# Shuffle the documents for randomness
random.shuffle(documents)

# Step 2: Data Preprocessing
# We'll just use the words as-is for this simple example
# You can use stop word removal, stemming (as shown above) to clean the data some more
documents = [(" ".join(words), category) for words, category in documents]

# Split into training and testing sets
train_data, test_data = train_test_split(documents, test_size=0.25, random_state=42)

# Step 3: Feature Extraction
# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the training data
X_train = vectorizer.fit_transform([review for review, _ in train_data])
y_train = [category for _, category in train_data]

# Transform the test data
X_test = vectorizer.transform([review for review, _ in test_data])
y_test = [category for _, category in test_data]

# Step 4: Model Building
# Initialize Naive Bayes Classifier and fit the data
classifier = MultinomialNB()
classifier.fit(X_train, y_train)

# Step 5: Evaluation
# Predict and evaluate the model
y_pred = classifier.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```
